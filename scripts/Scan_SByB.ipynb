```python
def scan_SByB_over_H1D(hSig, hBkgTot, xAxisThrshType, SEffiThrshs=[0.8, 0.6, 0.4]):
    #print(f"scan_SByB_over_H1D():: {xAxisThrshType = }  \nhSig.axes[0].centers ({len(hSig.axes[0].centers)}): {hSig.axes[0].centers} \nhSig.axes[0].edges ({len(hSig.axes[0].edges)}): {hSig.axes[0].edges} ")
    #print(f"hSig.values() ({len(hSig.values())}) ({type(hSig.values())}): {hSig.values()} ")
    #print(f"hBkgTot.values() ({len(hBkgTot.values())}) ({type(hBkgTot.values())}): {hBkgTot.values()} ")

    xBinFix = (len(hSig.axes[0].edges) - 1) if xAxisThrshType == ThrshTypes.LowerThrsh else 0
    #print(f"{xBinFix = }")
    
    #print(f"hSig.axes[0] ({type(hSig.axes[0])}): {hSig.axes[0]}")
    

    hSByB  = hist.Hist(hSig.axes[0])
    hSEffi = hist.Hist(hSig.axes[0])

    #print(f"hSByB ({type(hSByB)}): {hSByB}")
    #print(f"hSByB.values() ({len(hSByB.values())}) ({type(hSByB.values())}): {hSByB.values()} ")
    
    S_total = np.sum(hSig   .values())
    for xBinFloat in range(len(hSig.axes[0].edges) - 1):
        xBinLower = min(xBinFloat, xBinFix)
        xBinUpper = max(xBinFloat, xBinFix)
        S = np.sum(hSig   .values()[xBinLower : xBinUpper])
        B = np.sum(hBkgTot.values()[xBinLower : xBinUpper])
        SByB  = S / math.sqrt(B) if B > 0 else 0
        SEffi = S / S_total
        #print(f"x:{hSig.axes[0].centers[xBinFloat]}, xBinFloat: {xBinFloat}, [{xBinLower}, {xBinUpper}], S: {S} ({hSig.values()[xBinLower : xBinUpper].shape}) {hSig.values()[xBinLower : xBinUpper]}, B: {B}  ({hBkgTot.values()[xBinLower : xBinUpper].shape}), SByB: {SByB} ")
        #print(f" {hSig.axes[0].centers[xBinFloat]} ({xBinFloat}) {S = }, {B = }, {SByB = }")
        hSByB[xBinFloat] = SByB 
        hSEffi[xBinFloat] = SEffi

    bin_SByBMax = np.argmax( hSByB.values() )
    x_SByBMax = hSByB.axes[0].centers[bin_SByBMax]
    SByB_max = hSByB.values()[bin_SByBMax]

    x_SEffi_map = {}
    x_SEffi_SByB_map = {}
    for SEffiThrsh in SEffiThrshs:
        if xAxisThrshType == ThrshTypes.LowerThrsh:
            bin_SEffiPt = np.argmax( hSEffi.values() < SEffiThrsh ) # return aidx for first true
        else:
            bin_SEffiPt = np.argmax( hSEffi.values() > SEffiThrsh )
        x_SEffiPt = hSEffi.axes[0].centers[bin_SEffiPt]
        SByB_SEffiPt = hSByB.values()[bin_SEffiPt]
        x_SEffi_map[SEffiThrsh] = x_SEffiPt
        x_SEffi_SByB_map[x_SEffiPt] = [SEffiThrsh, SByB_SEffiPt]
        #print(f" hSEffi.values() < SEffiThrsh: { hSEffi.values() < SEffiThrsh}")
        #print(f"{bin_SEffiPt = }, {hSEffi.axes[0].centers[bin_SEffiPt] = }")
    
    #print(f"hSByB ({type(hSByB)}): {hSByB}")
    #print(f"hSByB.values() ({len(hSByB.values())}) ({type(hSByB.values())}): {hSByB.values()} ")

    #print(f"SByBMax: {np.max(hSByB.values())}, {bin_SByBMax = }, {x_SByBMax = }")

    return hSByB, SByB_max, x_SByBMax, hSEffi, x_SEffi_map, x_SEffi_SByB_map


```

```python
def scan_SByB_over_H2D(hSig, hBkgTot, xAxisThrshType, yAxisThrshType):
    if len(hSig.axes) != 2:
        print(f"scan_SByB_over_H2D():: len(hSig.axes) != 2 \t\t **** ERROR **** \nAborting...\n")
        exit(0)

    #print(f"scan_SByB_over_H2D():: {xAxisThrshType = }  \nhSig.axes[0].centers ({len(hSig.axes[0].centers)}): {hSig.axes[0].centers} \nhSig.axes[0].edges ({len(hSig.axes[0].edges)}): {hSig.axes[0].edges} ")
    #print(f"\nhSig.axes[1].centers ({len(hSig.axes[1].centers)}): {hSig.axes[1].centers} \nhSig.axes[1].edges ({len(hSig.axes[1].edges)}): {hSig.axes[1].edges} ")
    #print(f"hSig.values() ({len(hSig.values())}) ({type(hSig.values())}): {hSig.values().tolist()} ")
    #print(f"hBkgTot.values() ({len(hBkgTot.values())}) ({type(hBkgTot.values())}): {hBkgTot.values()} ")

    xBinFix = (len(hSig.axes[0].edges) - 1) if xAxisThrshType == ThrshTypes.LowerThrsh else 0
    yBinFix = (len(hSig.axes[1].edges) - 1) if xAxisThrshType == ThrshTypes.LowerThrsh else 0
    #print(f"{xBinFix = }, {yBinFix = }")
    
    #print(f"hSig.axes[0] ({type(hSig.axes[0])}): {hSig.axes[0]}")
    

    hSByB = hist.Hist(hSig.axes[0], hSig.axes[1])

    #print(f"hSByB ({type(hSByB)}): {hSByB}")
    #print(f"hSByB.values() ({len(hSByB.values())}) ({type(hSByB.values())}): {hSByB.values()} ")
    

    for xBinFloat in range(len(hSig.axes[0].edges) - 1):        
        xBinLower = min(xBinFloat, xBinFix)
        xBinUpper = max(xBinFloat, xBinFix)

        for yBinFloat in range(len(hSig.axes[1].edges) - 1):        
            yBinLower = min(yBinFloat, yBinFix)
            yBinUpper = max(yBinFloat, yBinFix)

            S = np.sum(hSig   .values()[xBinLower : xBinUpper, yBinLower : yBinUpper])
            B = np.sum(hBkgTot.values()[xBinLower : xBinUpper, yBinLower : yBinUpper])
            SByB = S / math.sqrt(B) if B > 0 else 0
            #print(f"(x, y): ({hSig.axes[0].centers[xBinFloat]}, {hSig.axes[1].centers[yBinFloat]}), BinFloat (x, y): ({xBinFloat}, {yBinFloat}), [{xBinLower} : {xBinUpper}, {yBinLower} : {yBinUpper}], S: {S} ({hSig.values()[xBinLower : xBinUpper, yBinLower : yBinUpper].shape}) {hSig.values()[xBinLower : xBinUpper, yBinLower : yBinUpper]}, B: {B}  ({hBkgTot.values()[xBinLower : xBinUpper, yBinLower : yBinUpper].shape}), SByB: {SByB} ")
            hSByB[xBinFloat, yBinFloat] = SByB 

    #print(f"\n\n\hSig.values() ({hSig.values().shape}):  {hSig.values().tolist() } ")
    #print(f"\n\n\nhSByB.values() ({hSByB.values().shape}):  {hSByB.values().tolist() } ")
    #print(f"\n\n\n {np.max(hSByB.values()) = },  {np.argmax(hSByB.values()) = },  {np.argmax(hSByB.values(), axis=0) = },  {np.argmax(hSByB.values(), axis=1) = } \n")

    #print(f"{hSByB.values().shape[0] = }, {hSByB.values().shape[1] = }, ")

    bin_SByBMax = np.argmax( hSByB.values() )
    xBin_SByBMax = int(bin_SByBMax / hSByB.values().shape[1])
    yBin_SByBMax = int(bin_SByBMax % hSByB.values().shape[1])
    SByB_max = hSByB.values()[xBin_SByBMax, yBin_SByBMax]
    #print(f"{bin_SByBMax = }, {xBin_SByBMax = }, {yBin_SByBMax = }, ")
    #print(f"{hSByB.values()[xBin_SByBMax, yBin_SByBMax] = }")


    x_SByBMax = hSByB.axes[0].centers[xBin_SByBMax]
    y_SByBMax = hSByB.axes[1].centers[yBin_SByBMax]
    
    #print(f"hSByB ({type(hSByB)}): {hSByB}")
    #print(f"hSByB.values() ({len(hSByB.values())}) ({type(hSByB.values())}): {hSByB.values()} ")

    #print(f"SByBMax: {np.max(hSByB.values())}, {bin_SByBMax = }, {xBin_SByBMax = }, {yBin_SByBMax = }, {x_SByBMax = }, {y_SByBMax = }")

    return hSByB, SByB_max, (x_SByBMax, y_SByBMax)
```
